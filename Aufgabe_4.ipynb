```python
#Explorativer Versuch
df2_max_play = df2.loc[df2.groupby('winner_name')['winner_total_games_played'].idxmax().tail(20)]
px.scatter_3d(df2_max_play, x='winner_name', y='winner_age', z='winner_win_rate')
```

```python
#Explorativer Versuch
df_agg = df2.groupby(['winner_name','winner_age'], as_index=False)[['winner_total_games_played']].max()
payload = pd.merge(df2, df_agg)[['winner_name','winner_total_games_played','winner_win_rate','winner_age']].tail(1000)
px.line(payload, x='winner_age', y='winner_win_rate', color='winner_name').update_layout(showlegend=False)
```

```python
#Explorativer Versuch
df_agg = df2.groupby(['winner_age','winner_name'], as_index=False)[['winner_win_rate']].mean()
payload = pd.merge(df2, df_agg).tail(1000)#[['winner_name','winner_total_games_played','winner_win_rate','winner_age']].tail(1000)
px.line(payload, x='winner_age', y='winner_win_rate', color='winner_name')#.update_layout(showlegend=False)
```

```python
df_agg = df2.groupby(['winner_name','winner_age'])[['winner_win_rate']].mean()
proj = pd.merge(df2, df_agg).tail(1000)
px.line(proj, x='winner_age', y='winner_win_rate', color='winner_name')
```

```python
df_agg = df2[(df2.winner_name=='Rafael Nadal')]
proj = pd.merge(df2, df_agg)
px.line(proj, x='winner_age', y='winner_win_rate', color='winner_name', title='Raphael Nadal').update_layout(showlegend=False)
```

```python
df_agg = df2[(df2.winner_name=='Boris Becker')]
proj = pd.merge(df2, df_agg)
px.line(proj, x='winner_age', y='winner_win_rate', color='winner_name', title='Boris Becker').update_layout(showlegend=False)
```
